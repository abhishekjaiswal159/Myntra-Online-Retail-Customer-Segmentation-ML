## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# üîß Step 1: Remove duplicate rows
df.drop_duplicates(inplace=True)

# üßπ Step 2: Drop rows with missing critical information (e.g., InvoiceNo, StockCode, Description, CustomerID)
df.dropna(subset=['InvoiceNo', 'StockCode', 'Description', 'CustomerID'], inplace=True)

# üîÑ Step 3: Convert InvoiceNo and StockCode to string (they are IDs, not numerical values)
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
df['StockCode'] = df['StockCode'].astype(str)

# üïí Step 4: Convert InvoiceDate to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# üí∏ Step 5: Remove rows with negative or zero UnitPrice or Quantity
df = df[(df['UnitPrice'] > 0) & (df['Quantity'] > 0)]

# ‚úÖ Step 6: Reset index after cleaning
df.reset_index(drop=True, inplace=True)

# üì§ Step 7: Assign cleaned data to a new variable for analysis
df_cleaned = df.copy()

# üîÅ Final check after cleaning
print("‚úÖ Cleaned Dataset Shape:", df.shape)
print("‚úÖ Any missing values left?\n", df.isnull().sum())
````

#### Chart - 4

````python
# ‚úÖ Step 1: Create 'Revenue' column
df_cleaned['Revenue'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']

# ‚úÖ Step 2: Extract Year-Month for grouping
df_cleaned['InvoiceMonth'] = df_cleaned['InvoiceDate'].dt.to_period('M')

# ‚úÖ Step 3: Calculate monthly revenue
monthly_revenue = df_cleaned.groupby('InvoiceMonth')['Revenue'].sum().reset_index()

# ‚úÖ Step 4: Convert InvoiceMonth to string to avoid plotting issues
monthly_revenue['InvoiceMonth'] = monthly_revenue['InvoiceMonth'].astype(str)

# ‚úÖ Step 5: Plotting monthly revenue trend
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_revenue, x='InvoiceMonth', y='Revenue', marker='o', color='teal')
plt.title('üìà Monthly Revenue Trend Over Time', fontsize=16)
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue (¬£)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

We selected a **line chart** for this analysis because line plots are ideal for visualizing **trends over time**. Since we wanted to analyze how the revenue changes from month to month, this type of chart helps us:

- Observe the **overall revenue trend** over the year.
- Identify **peak months** and **slow periods** in business performance.

A line chart is simple yet powerful in displaying time-series data, which is why it is most suitable for this insight.

Using a heatmap helps us:
- Quickly assess which numerical features are positively or negatively correlated.
- Identify potential multicollinearity, which is crucial for future predictive modeling.
- Understand how pricing and quantity sold might be interacting with each other.
