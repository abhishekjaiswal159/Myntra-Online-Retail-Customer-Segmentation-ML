## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# 🔧 Step 1: Remove duplicate rows
df.drop_duplicates(inplace=True)

# 🧹 Step 2: Drop rows with missing critical information (e.g., InvoiceNo, StockCode, Description, CustomerID)
df.dropna(subset=['InvoiceNo', 'StockCode', 'Description', 'CustomerID'], inplace=True)

# 🔄 Step 3: Convert InvoiceNo and StockCode to string (they are IDs, not numerical values)
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
df['StockCode'] = df['StockCode'].astype(str)

# 🕒 Step 4: Convert InvoiceDate to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 💸 Step 5: Remove rows with negative or zero UnitPrice or Quantity
df = df[(df['UnitPrice'] > 0) & (df['Quantity'] > 0)]

# ✅ Step 6: Reset index after cleaning
df.reset_index(drop=True, inplace=True)

# 📤 Step 7: Assign cleaned data to a new variable for analysis
df_cleaned = df.copy()

# 🔁 Final check after cleaning
print("✅ Cleaned Dataset Shape:", df.shape)
print("✅ Any missing values left?\n", df.isnull().sum())
````

## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
# 📦 Importing necessary libraries
import pandas as pd

# 🔍 Check missing values in the dataset
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_percent = (missing_values / len(df)) * 100

# 📊 Display columns with missing values and their percentage
missing_data = pd.DataFrame({
    'Missing Values': missing_values,
    'Percentage (%)': missing_percent
})
missing_data = missing_data[missing_data['Missing Values'] > 0]
print("📉 Missing Values Summary:")
display(missing_data)

# 🛠️ Imputation strategy
# Impute numerical columns with median and categorical columns with mode
for column in df.columns:
    if df[column].isnull().sum() > 0:
        if df[column].dtype in ['int64', 'float64']:
            median_value = df[column].median()
            df[column].fillna(median_value, inplace=True)
        else:
            mode_value = df[column].mode()[0]
            df[column].fillna(mode_value, inplace=True)

# ✅ Check again to ensure no missing values
print("✅ Are there any missing values left?")
print(df.isnull().sum().any())
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

We have applied the following missing value imputation techniques:

1. **Median Imputation for Numerical Columns**:
   - **Why?** Median is robust to outliers and gives a better central tendency measure when the data is skewed.
   - **When?** Used for features with numerical data types (`int` or `float`) that had missing values.

2. **Mode Imputation for Categorical Columns**:
   - **Why?** Mode represents the most frequent category and helps maintain the distribution of categorical variables.
   - **When?** Used for features with object or categorical data types.

These techniques were selected because:
- They are simple, interpretable, and effective for handling moderate amounts of missing data.
- They avoid introducing significant bias into the dataset, especially in the absence of complex relationships or patterns in the missingness.

By applying these methods, we preserved the integrity of the dataset and ensured it is clean for further analysis and modeling.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments# 📦 Import necessary library
import numpy as np

# 📊 Select numerical columns for outlier treatment
numerical_cols = df.select_dtypes(include=np.number).columns

# 🔧 Handle outliers using the IQR method
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

# ✅ Display confirmation
print("Outliers have been handled using IQR capping technique.")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

We have used the following outlier treatment techniques in our analysis:

1. **Interquartile Range (IQR) Method**:
   - **What it does**: Identifies outliers as values below `Q1 - 1.5 × IQR` or above `Q3 + 1.5 × IQR`.
   - **Why used**: It is a robust method that is not affected by extreme values or skewed distributions. It helped us detect and filter unrealistic high or low values without affecting the core data distribution.

2. **Winsorization (Capping)**:
   - **What it does**: Limits extreme values in the dataset by capping them at specific percentiles (e.g., 1st and 99th percentile).
   - **Why used**: This retains all data points but reduces the influence of extreme outliers, which is useful for models sensitive to scale or spread of values.

3. **Log Transformation** *(if required)*:
   - **What it does**: Applies a logarithmic scale to highly skewed numerical data.
   - **Why used**: To reduce the effect of extreme values, normalize distributions, and make patterns more visible during visualization and correlation analysis.

These methods helped us ensure that the data is clean, the range is balanced, and insights derived are not misleading due to anomalies.

#### 6. Rephrase Text

````python
# Rephrase Text
# 📚 Import library
from nltk.corpus import wordnet
import nltk
nltk.download('wordnet')

# 🔁 Simple synonym replacement
def simple_rephrase(text):
    if not isinstance(text, str):
        return text

    words = text.split()
    new_words = []

    for word in words:
        syns = wordnet.synsets(word)
        if syns:
            synonym = syns[0].lemmas()[0].name()
            new_words.append(synonym)
        else:
            new_words.append(word)

    return ' '.join(new_words)

# 🧾 Apply to relevant column
for col in text_columns:
    if col in df.columns:
        df[col] = df[col].apply(simple_rephrase)

# ✅ Done
print("Text rephrased using simple synonym replacement.")
````

### TF-IDF Vectorization for Lemmatized Text Reviews

We have used the **TF-IDF (Term Frequency-Inverse Document Frequency)** vectorization technique for transforming the lemmatized text reviews into numerical format.

---

#### ✅ Why TF-IDF?

TF-IDF is ideal for our use case because:

1. **Balances Common vs. Rare Words**  
   It gives higher importance to rare but informative words (e.g., *“glitch”*, *“laggy”*) and less weight to very common words (e.g., *“the”*, *“app”*) that appear in almost every review.

2. **Better Feature Weighting**  
   Unlike simple count-based methods (like Bag of Words), TF-IDF adjusts for how frequently a term appears across all documents, helping highlight keywords that distinguish reviews.

3. **Reduces Noise**  
   Words that are common across all reviews (e.g., *“good”*, *“app”*) are down-weighted, making it easier to spot patterns for tasks like sentiment analysis or clustering.

4. **Sparse & Efficient**  
   Produces a sparse matrix, which is computationally efficient for large datasets.

---

#### ⚙️ Formula Recap

For each term *t* in a document *d*:

- **TF (Term Frequency)** = How often *t* appears in *d*  
- **IDF (Inverse Document Frequency)** = log(Total Documents / Documents containing *t*)  
- **TF-IDF** = TF × IDF  

---

#### 🧠 Summary

We chose **TF-IDF** because it strikes a balance between simplicity and smart weighting, making it ideal for extracting important terms from text reviews that influence app sentiment or usefulness.

---

**3. Recall**
- **What it indicates:**  
  Out of all actual instances of a class, how many the model correctly identified.
- **Business Impact:**  
  High recall is crucial when missing out on important cases is costly. For instance, when identifying dissatisfied customers, high recall ensures most unhappy customers are captured for retention efforts.

---

**4. F1-Score**
- **What it indicates:**  
  The harmonic mean of precision and recall. A balanced metric for uneven class distributions.
- **Business Impact:**  
  F1-score gives a single measure of model performance, especially when there’s a trade-off between precision and recall. It's useful when businesses want to optimize both customer retention and targeting accuracy.

---

**5. Confusion Matrix**
- **What it indicates:**  
  A summary of prediction errors per class.
- **Business Impact:**  
  Helps stakeholders see which specific customer segments or product categories are being misclassified, allowing them to take corrective actions like retraining models or collecting more data.
